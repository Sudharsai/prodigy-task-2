```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_data = pd.read_csv(r'C:\Users\Sudarsan\OneDrive\sudarsun\data set\task 2 data.csv', usecols=['PassengerId', 'Survived'])
```

```python
titanic_data
```

```python
print("Number of missing values in each column:")
print(titanic_data.isnull().sum())
```

```python
titanic_data = titanic_data.dropna()
```

```python
print("\nSurvival counts:")
print(titanic_data['Survived'].value_counts())
```

```python
plt.figure(figsize=(4, 5))
sns.countplot(x='Survived', data=titanic_data)
plt.title('Distribution of Survival')
plt.xlabel('Survival (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='PassengerId', y='Survived', data=titanic_data)
plt.title('Relationship between Passenger ID and Survival')
plt.xlabel('Passenger ID')
plt.ylabel('Survival (0 = No, 1 = Yes)')
plt.show()
```
